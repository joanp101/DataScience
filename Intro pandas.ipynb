# gruopby

```python
df.groupby(by = "DayOfWeek")["ArrDelay"].max() #por cadia dia de la semana ("DayOfWeek") el Retraso ("ArrDelay") maximo de vuelos 
```

```python
df.groupby(by = "DayOfWeek")["ArrDelay"].describe() #resumen estadistico agrupado para cada uno de los dias de la semana
```

```python
df.groupby(by = "DayOfWeek")["ArrDelay"].mean()# Calcula la media para la columna ArrDelay
```

```python
df.groupby(by = "DayOfWeek")["ArrDelay", "DepDelay"].mean()# Calcula la media a dos columnas
```

```python
df.groupby(by = "DayOfWeek")["ArrDelay"].max()-df.groupby(by = "DayOfWeek")["ArrDelay"].min()#operacion de resta para max y min(rango de valores)
```

```python
dfATLHOU=df[df.Origin.isin(["ATL","HOU"])]#crear nuevo dataframe con vuelos de dos ciudades ATLANTA y HOUSTONG
```

```python
#por cada dia de la semana aquellos vuelos que salen de atlanta y houstom y sus medias
#lo que permite comparar cual de las dos ciudades cuales vuelos llegan mas retrasados y en que dia las diferencias son mayores
dfATLHOU.groupby(by=["DayOfWeek", "Origin"])["ArrDelay"].mean()
```
